**Condicionales (if)**

Se dividen en bloques:
if condición1:   (solo una vez dentro de un bloque)
  bloque código
elif condición2: (similar a un if anidado en Excel, las veces que sea necesario, mientras aguante la memoria RAM)
  bloque código
...
else:            (opcional)
  bloque código


Python valida las condiciones de arriba hacia abajo. Donde llegue a validar Python, allí se detiene. OJO con los indentados o espaciados dentro del código.

Condiciones a utilizar dentro de if:
- Equals: a == b
- Not equals: a != b
- Less than: a < b
- Less than or equal to: a <= b
- Greater than: a > b
- Greater than or equal to: a >= b

**Ejercicio 4 - CONDICIONALES:**
Una tienda vende un producto a precios unitarios que dependen de la cantidad de unidades adquiridas de acuerdo a la tabla adjunta:

de 1 a 25: precio 27.7
de 26 a 50: precio 25.5
de 51 a 75: precio 23.7
de 76 a más : precio 21.7
Adicionalmente, si el cliente adquiere más de 50 unidades la tienda le descuenta el 15% del importe de la compra; en caso contrario, sólo le descuenta el 5% del importe compra.
Diseñe un programa que determine el importe de la compra, el importe del descuento y el importe a pagar por la compra de cierta cantidad de unidades del producto.

```python
cant = int(input("Ingrese la cantidad de unidades a adquirir: "))

if 1 < cant <= 25:
  pu = 27.7
elif 26 <= cant <= 50:
  pu = 25.5
elif 51 <= cant <= 75:
  pu = 23.7
elif cant >= 76:
  pu = 21.7
print("El precio unitario por las", cant, "adquiridas es de: ", pu)

subtot = cant * pu
print("El subtotal de la compra de", cant, "unidades, es de: ", subtot)

if cant > 50:
  descto = 0.15
else:
  descto = 0.05
print("El descuento asignado es de: ", descto * 100, "%", sep = "")

total = round(subtot * (1-descto) ,2)
print("El importe total es de: ", total)
```

# Copia de listas
Existen dos formas de copiar listas:

* Copia por referencia l1 = l2: Asocia la la variable l1 la misma lista que tiene asociada la variable l2, es decir, ambas variables apuntan a la misma dirección de memoria. Cualquier cambio que hagamos a través de l1 o l2 afectará a la misma lista.

* Copia por valor l1 = list(l2): Crea una copia de la lista asociada a l2 en una dirección de memoria diferente y se la asocia a l1. Las variables apuntan a direcciones de memoria diferentes que contienen los mismos datos. Cualquier cambio que hagamos a través de l1 no afectará a la lista de l2 y viceversa.

```python
# Copiar referencia / espejo (lo que le pase a una lista, le pasa a la otra):
listR1 = [1, 2, 3, 4, 5]
listR1
```

**II. Tuplas**
Una tupla es una secuencia ordenada de objetos de distintos tipos.

**Muy similares a un diccionario, con la diferencia de que NO SON MUTABLES.**

Se construyen poniendo los elementos entre paréntesis ( ) separados por comas.

Se caracterizan por:
